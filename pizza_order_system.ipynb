```python
# Import libraries
import csv
from datetime import datetime
import pandas as pd
```

```python
# Superclass
class Pizza:        
    def __init__(self):
        self.description = "Pizza"
        self.cost = 0.00
        
    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost

# Subclass
class ClassicPizza(Pizza):
    def __init__(self):
        self.description = "Classic Pizza"
        self.cost = 10.00

class MargaritaPizza(Pizza):
    def __init__(self):
        self.description = "Margarita Pizza"
        self.cost = 15.00

class TurkPizza(Pizza):
    def __init__(self):
        self.description = "Turk Pizza"
        self.cost = 17.00
        
class SadePizza(Pizza):
    def __init__(self):
        self.description = "Sade Pizza"
        self.cost = 7.00
        
# Superclass
class Decorator:        
    def __init__(self):
        self.description = "sauce"
        self.cost = 0.00
        
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
    
class Onion(Decorator):
    def __init__(self):
        self.description = "Onions"
        self.cost = 10.00
    
class Mashrooms(Decorator):
    def __init__(self):
        self.description = "Mashrooms"
        self.cost = 10.00

class Cheese(Decorator):
    def __init__(self):
        self.description = "Cheese"
        self.cost = 19.00
    
class Meat(Decorator):
    def __init__(self):
        self.description = "Meat"
        self.cost = 30.00
    
class Corn(Decorator):
    def __init__(self):
        self.description = "Corn"
        self.cost = 13.00
    
class Olives(Decorator):
    def __init__(self):
        self.description = "Olives"
        self.cost = 5.00
```

```python
# Main function
def main():
    menu = pd.read_csv("data/menu.txt")
    print(menu)
    
    pizzaChoice = input("Please write the number of the pizza you want to choose.\n")
    
    # Choose pizza base.
    if pizzaChoice == '1':
        Pizza = ClassicPizza()
    elif pizzaChoice == '2':
        Pizza = MargaritaPizza()
    elif pizzaChoice == '3':
        Pizza = TurkPizza()
    elif pizzaChoice == '4':
        Pizza = SadePizza()
    else:
        print("Wrong number! Please try again.")
        return
    
    # Choose sauces.
    saucechoice = []
    print("Please write the number of the souce you want to choose. After your sauce selection is done, you can type 'done'.\n")
    while True:
        sauceChoice = input("Please write the number of the souce you want to choose --> ")    
    
        if sauceChoice == '11':
            sauce = Olives()
        elif sauceChoice == '12':
            sauce = Mashrooms()
        elif sauceChoice == '13':
            sauce = Cheese()
        elif sauceChoice == '14':
            sauce = Meat()
        elif sauceChoice == '15':
            sauce = Onion()
        elif sauceChoice == '16':
            sauce = Corn()
        elif sauceChoice == 'done':
            break
        else:
            print("Wrong number! Please try again.")
        saucechoice.append(sauce)
    
    # Order Time
    orderTime = datetime.now().strftime("%d-%m-%Y // %H:%M:%S")
    
    # Get order's total cost and total description.
    TotalCost = Pizza.get_cost()
    OrderDescription = Pizza.get_description()
    
    for sauce in saucechoice:
        TotalCost += sauce.get_cost()
        OrderDescription += ", " + sauce.get_description()
    
    # Get user's information
    print("\n**********")
    Name= input("Please enter your name: ")
    UserID= input("Please enter your ID: ")
    CreditCardNumber= input("Please enter your credit card number: ")
    CreditCardPword= input("Please enter your credit card password: ")
    print("**********")
    print("Thank you for your order!\n")

    # Open .csv file
    with open("Orders_Database.csv", mode = "a", newline="") as f:
        writer = csv.writer(f)

        # If file is empty, print the header line.
        if f.tell() == 0:
            f.write("Name"+","+"User ID"+","+"Credit Card Number"+","+"Total Cost"+","+"Order Description"+","+"Order Time"+","+"Credit Card Password\n")

        # User information
        writer.writerow([Name, UserID, CreditCardNumber, TotalCost, OrderDescription, orderTime, CreditCardPword])

    # Close the file
    f.close()
```

```python
# Run main code.
main()
```
